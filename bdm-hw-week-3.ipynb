# Ex 4. Randomly create a data matrix of 10 samples with 5 features. 
---------------------------------------------------------------------
* Then, randomly create a label vector for 10 samples. 
* Finally, seperate samples into 2 groups
* Group 1 called train_data (label == 0)
* Group 2 called test_data (label == 1)


```python
rng = np.random.RandomState(0)
main_matrix = rng.randint(10, size = (10,5))
main_matrix
```

```python
rng = np.random.RandomState(0)
data_matrix = rng.randint(2 ,size = 10)
data_matrix
```

```python
train_data = main_matrix[np.where(data_matrix == 1)]
train_data
```

```python
test_data = main_matrix[np.where(data_matrix == 0)]
```
